```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/content/fruit.jpg')

plt.figure(figsize=(10, 10))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Original')

r, c = image.shape[0:2]

for i in range(r):
  for j in range(c):
    image[i, j] = sum(image[i, j]) * 0.33

plt.subplot(1,2,2)
plt.imshow(image)
plt.title('Gray Image')
```
