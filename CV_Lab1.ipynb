```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/high_contrast.jpg')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

s_f_1 = 5.0
s_f_2 = 1/5.0

height, width = image_rgb.shape[:2]

new_height = int(height * s_f_1)
new_width = int(width * s_f_1)

zoomed_image = cv2.resize(src=image_rgb, dsize=(new_height,new_width), interpolation=cv2.INTER_CUBIC)

new_height1 = int(height * s_f_2)
new_width1 = int(width * s_f_2)

scaled_image = cv2.resize(src=image_rgb,dsize=(new_width1,new_height1),interpolation=cv2.INTER_AREA)

fig, axis = fig,axs=plt.subplots(1,3,figsize=(10,4))
```

```python


# Display the image
plt.imshow(image_rgb)
axs[0].set_title("Original Image Shape: " + str(image_rgb.shape))
plt.show()



plt.imshow(zoomed_image)
axs[1].set_title("Zoomed Image Shape:"+str(zoomed_image.shape))
axs[1].axis('off')
plt.show()

plt.imshow(scaled_image)
axs[2].set_title("Scaled Image Shape:"+str(scaled_image.shape))
axs[2].axis('off')


# Show the plot
plt.show()
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

center=(image_rgb.shape[1]//2,image_rgb.shape[0]//2)
print("center of image",center)
angle=90
scale=1


rotation_matrix=cv2.getRotationMatrix2D(center,angle,scale)


rotated_image=cv2.warpAffine(image_rgb,rotation_matrix,(image.shape[1],image.shape[0]))


fig,axs=plt.subplots(1,2,figsize=(7,4))


axs[0].imshow(image_rgb)
axs[0].set_title("Original Image")



axs[1].imshow(rotated_image)
axs[1].set_title("Rotated Image")

plt.show()
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)



width=image_rgb.shape[1]
height=image_rgb.shape[0]


tx=50
ty=70

t_m=np.array([[1,0,tx],[0,1,ty]],dtype=np.float32)
t_i=cv2.warpAffine(image_rgb,t_m,(width,height))
fig,axs=plt.subplots(1,2,figsize=(7,4))


axs[0].imshow(image_rgb)
axs[0].set_title("Original Image")



axs[1].imshow(t_i)
axs[1].set_title("Translated Image")

plt.show()
```
