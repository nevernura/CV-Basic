```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Function for erosion
def erosion(img, SE):
    imgErode = cv2.erode(img, SE, iterations=1)
    return imgErode

# Function for dilation
def dilation(img, SE):
    imgDilate = cv2.dilate(img, SE, iterations=1)
    return imgDilate

# Read the image for dilation
img = cv2.imread('/content/OIP.jpg', 0)  # Load as grayscale

# Define the structuring element using inbuilt CV2 function
SE = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

# Dilate the image
AdB = dilation(img, SE)

# Erode the dilated image. This gives the closing operation
CdAB = erosion(AdB, SE)

# Plot all the images
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(2, 2, 2)
plt.imshow(AdB, cmap='gray')
plt.title('D(A, B)')

plt.subplot(2, 2, 3)
plt.imshow(CdAB, cmap='gray')
plt.title('C(D(A,B), B)')

plt.tight_layout()
plt.show()
```
