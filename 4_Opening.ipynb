```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Function for erosion
def erosion(img, SE):
    imgErode = cv2.erode(img, SE, iterations=1)
    return imgErode

# Function for dilation
def dilation(img, SE):
    imgDilate = cv2.dilate(img, SE, iterations=1)
    return imgDilate

# Read the image for dilation
img = cv2.imread('/content/OIP.jpg')

# Define the structuring element using inbuilt CV2 function
SE = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

# Erode the image
AeB = erosion(img, SE)

# Dilate the eroded image. This gives the opening operation
AoB = dilation(AeB, SE)

# Opening - Erosion then Dilation
AoBdB = erosion(AoB, SE)
AoBdBeB = dilation(AoBdB, SE)

# Plot all the images
plt.figure(figsize=(10, 10))

plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(3, 2, 2)
plt.imshow(AeB, cmap='gray')
plt.title('E(A, B)')

plt.subplot(3, 2, 3)
plt.imshow(AoB, cmap='gray')
plt.title('O(A, B)')

plt.subplot(3, 2, 4)
plt.imshow(AoBdB, cmap='gray')
plt.title('D(O(A,B), B)')

plt.tight_layout()
plt.show()
```
