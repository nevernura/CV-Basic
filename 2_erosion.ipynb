```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/pexels-vitor-gusmao-shimabukuro-1110669-2104044.jpg', 0)

#Acquire size of image
m, n = img1.shape

#Define of structuring element
k = 15
SE = np.zeros((k, k), dtype=np.uint8)
constant = (k - 1) // 2

#Define new image
imgErode = np.zeros((m, n), dtype=np.uint8)

#Erosion without library
for i in range(constant, m-constant):
  for j in range(constant, n - constant):
    temp = img1[i - constant: i+constant+1, j-constant: j+constant+1]
    product = temp * SE
    imgErode[i, j] = np.min(product)

#Plotting result
plt.figure(figsize=(10, 10))

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Original')
plt.imshow(imgErode, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
input = cv2.imread(f'/content/OIP.jpg')

plt.figure(figsize=(5, 5))
plt.subplot(1,2,1)
plt.imshow(input)
plt.title('Original')

kernel = np.ones((3,3), np.uint8)

eroded = cv2.erode(input, kernel, iterations=1)

plt.figure(figsize=(5, 5))
plt.subplot(1,2,2)
plt.imshow(eroded)
plt.title('Eroded Image')
```
