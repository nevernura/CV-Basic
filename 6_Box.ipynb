```python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/pexels-borja-lopez-2164683.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

scale_factor = 1/30

height, width = image_rgb.shape[:2]

new_height1 = int(height * scale_factor)
new_width1 = int(width * scale_factor)

# Scaled image
scaled_image = cv2.resize(src= image_rgb,
                          dsize =(new_width1, new_height1),
                          interpolation=cv2.INTER_AREA)


# Convert the image to grayscale
#gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define the kernel size (box size)
kernel_size = (18, 18)  # Adjust the kernel size as needed

# Apply the box filter
box_filtered_image = cv2.boxFilter(scaled_image, -1, kernel_size)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot the original image
axs[0].imshow(scaled_image)
axs[0].set_title('Original Image')

# Plot the Zoomed Image
axs[1].imshow(box_filtered_image)
axs[1].set_title('Box Filtered Image')

# Remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# Display the subplots
plt.tight_layout()
plt.show()
```
